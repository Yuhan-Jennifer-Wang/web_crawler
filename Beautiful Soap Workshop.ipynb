```python
# importing libraries
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.imdb.com/search/title?genres=drama&groups=top_250&sort=user_rating,desc"
res = requests.get(url) # response from website's server
```

```python
print(res)
```

```python
type(res)
```

```python
soup = BeautifulSoup(res.text, 'lxml')
print(soup)
```

```python
movie = soup.select_one(".lister-item-header a")
print(movie)

print(movie["href"])

print(movie.text)
```

# Selecting Multiple Elements

```python
movies = soup.select(".lister-item-header a")
```

```python
print(movies)
```

```python
print(movies[0])
```

```python
movies_titles = []
movies_links = []

for item in movies:
    movies_titles.append(item.text)
    link = "http://imdb.com" + item["href"]
    movies_links.append(link)

print(movies_titles)
print("\n")
print(movies_links)
```
